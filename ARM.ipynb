**Association Rule Mining**

Membuat pivot table untuk mengelompokan judul buku berdasarkan pembaca

```python
def create_pivot (dataframe, id=False):
    if id:
        return dataframe.groupby(['user_id', "book_id"])['count'].sum().unstack().fillna(0). \
            applymap(lambda x: 1 if x > 0 else 0)
    else:
        return dataframe.groupby(['user_id', 'book_title'])['count'].sum().unstack().fillna(0). \
            applymap(lambda x: 1 if x > 0 else 0)
```

```python
title = create_pivot(df_read)
title=title.drop(['11/22/63'], axis=1)
title
```

```python
frequent_itemsets = apriori(title,min_support=0.001,use_colnames=True)
frequent_itemsets
```

```python
associationrule = association_rules(frequent_itemsets,metric="support",min_threshold=0.0005)
associationrule
```

```python
# Menampilkan range value dari kolom support, confidence, dan lift dari Rules
print('Range Nilai Support')
print('Min:', associationrule['support'].min(), '\nMax:', associationrule['support'].max())
print('\nRange Nilai Confidence')
print('Min:', associationrule['confidence'].min(), '\nMax:', associationrule['confidence'].max())
print('\nRange Nilai Lift')
print('Min:', associationrule['lift'].min(), '\nMax:', associationrule['lift'].max())
```

```python
# Mengurutkan data berdasarkan nilai confidence tertinggi
associationrule.sort_values(by='confidence', ascending=False).head(25)
```

```python
print('\nRange Nilai Lift dengan Confidence == 100%')
print('Min:', (associationrule.sort_values(by='confidence', ascending=False).head(25))['lift'].min(), '\nMax:', (associationrule.sort_values(by='confidence', ascending=False).head(25))['lift'].max())
```

```python
# Rules dengan nilai confidence tertinggi
associationrule.sort_values(by='confidence', ascending=False).head(120)[associationrule['confidence'] >= 1]
```

```python
# Info Rules dengan nilai confidence tertinggi
associationrule.sort_values(by='confidence', ascending=False).head(120)[associationrule['confidence'] >= 1].info()
```

```python
# Rule dengan nilai confidence 100% dan lift tertinggi
associationrule.sort_values(by='confidence', ascending=False).head(25)[associationrule['lift'] >= 211.161]
```
